# Deepfake Detection Preprocessing - Local Development
Extract faces from videos and save them for training.

```python
import os
import glob
import cv2
import numpy as np
from tqdm import tqdm
from pathlib import Path

# Local configuration
DATASET_ROOT = Path("datasets")
OUTPUT_ROOT = Path("datasets/real_and_fake_face")

os.makedirs(OUTPUT_ROOT / "training_real", exist_ok=True)
os.makedirs(OUTPUT_ROOT / "training_fake", exist_ok=True)

print(f"Dataset Root: {DATASET_ROOT.absolute()}")
```

```python
def extract_faces_from_video(video_path, output_dir, max_frames=100):
    cap = cv2.VideoCapture(str(video_path))
    count = 0
    video_name = Path(video_path).stem
    
    # Note: For face detection locally, we recommend using cv2.CascadeClassifier 
    # or a dedicated library like face_recognition if installed.
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    while count < max_frames:
        ret, frame = cap.read()
        if not ret: break
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        
        for i, (x, y, w, h) in enumerate(faces):
            face_img = frame[y:y+h, x:x+w]
            face_img = cv2.resize(face_img, (224, 224))
            cv2.imwrite(str(output_dir / f"{video_name}_face_{count}_{i}.jpg"), face_img)
            
        count += 1
    cap.release()

print("Preprocessing function defined.")
```
